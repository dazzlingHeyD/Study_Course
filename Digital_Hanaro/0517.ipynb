## ✍️**딕셔너리 min, max정리 다시 하기 **

```python
# 딕셔너리 형태의 요소를 가지고 있는 리스트 
players = [
    {"이름" : "강백호",
     "신장": 189},
     {"이름" : "서태웅",
     "신장": 185},
     {"이름" : "채치수",
     "신장": 193}
]


def height(players):
  return players["신장"] #players['신장']
# max함수 안에서 
# key에서 사욜하고 있는 함수의 매개변수로 딕셔너리 players사용
print(min(players,key = height)) # 최솟값  # 189,185,193
print(max(players,key = height)) # 최댓값
```

```python
# 딕셔너리 형태의 요소를 가지고 있는 리스트 
players = [
    {"이름" : "강백호",
     "신장": 189},
     {"이름" : "서태웅",
     "신장": 185},
     {"이름" : "채치수",
     "신장": 193}
]



# max함수 안에서 
# key에서 사욜하고 있는 함수의 매개변수로 딕셔너리 players사용
print(min(players,key = lambda x:x["신장"])) # 최솟값  # 189,185,193
print(max(players,key = lambda x:x["신장"])) # 최댓값
```

### 텍스트 파일 한줄씩 읽기 

```python
hangul =  list("가나다라마바사아자차카타파하")
print(hangul)
```

```python
# 이름, 키, 몸무게 => csv텍스트 파일로 저장 
import random 
hangul =  list("가나다라마바사아자차카타파하")
with open('info.txt', "w") as file:
  for i in range(1000):
    name = random.choice(hangul)+random.choice(hangul)+random.choice(hangul)
    height = random.randrange(100,200) 
    weight = random.randrange(30,100)
    file.write("{},{},{}\n".format(name, height,weight))

```

```python
# 텍스트 파일 한줄씩 읽기 
with open('info.txt', "r") as file:
  for line in file:
    print(line)
```

```python
# 텍스트 파일 한 줄씩 읽기 & 새로운 변수 생성 및 출력 
with open('info.txt', "r") as file:
  for line in file:
      name, height, weight = line.strip().split(",")  #line.strip().split(",") 문자열 
      if (not name) or (not height) or (not weight): 
        continue
      
      # 체중(kg)/ 키(m)제곱
      bmi = int(weight) / (int(height)/100)**2 
      result = ""  #bmi 결과 넣을 문자열 변수 
      # 25보다 크거나 같으면 과체중
      # 18.5보다 크거나 같으면 정상 
      # 나머지는 저체중 
      if bmi>25:
        result = "과체중"
      elif bmi>=18.5:
        result = "정상"
      else: 
        result = "저체중"
      print("이름:{},키:{},몸무게:{},BMI{:.2f}, 결과:{}".format(name,height,weight,bmi,result))
```

### try except else

```python
# try - except pass
try: 
  num = input(" 원의 반지름 정수로 입력 >>")
  pi = 3.14
  num = int(num)
  print("원의 반지름:",num)
  print("원의 둘레:",2*pi*num)
  print("원의 넓이:",pi*num*num)
except:pass # 정상 종료가 됨 # 어디에서 오류가 났는지 모름 
```

```python
# try - except pass
try: 
  num = input(" 원의 반지름 정수로 입력 >>")
  pi = 3.14
  num = int(num)
except:
  print(" 정수입력하세요! ")
else:  # 정상 값 입력된 경우 
    print("원의 반지름:",num)
    print("원의 둘레:",2*pi*num)
    print("원의 넓이:",pi*num*num)
```

### try-except-else-finally 

```python
# try - except pass
try:  # 예외 발생할 수 있는 코드 입력 
  num = input(" 원의 반지름 정수로 입력 >>")
  pi = 3.14
  num = int(num)
except: #예외 발생할 때 처리할 코드 
  print(" 정수만 입력하세요! ")
else:  # 예외 발생하지 않는 정상 값 입력된 경우 
    print("원의 반지름:",num)
    print("원의 둘레:",2*pi*num)
    print("원의 넓이:",pi*num*num)
finally: 
  print("정상 종료")
```

## 예외객체 

```python
# 예외를 저장하는 객체 
try: 
  num = input(" 원의 반지름 정수로 입력 >>")
  pi = 3.14
  num = int(num)
  print("원의 반지름:",num)
  print("원의 둘레:",2*pi*num)
  print("원의 넓이:",pi*num*num)
except Exception as e:  # 예외 발생시키지 않고 e:예외 객체 
  print("예외 종류:", type(e))
  print("예외 객체:", e)
```

```python
# 여러가지 예외 상황(2가지)
# 1번째 예외 상황: 정수가 아닌 문자열 입력 
    # 예외 종류: <class 'ValueError'>, 
    # 예외 객체: invalid literal for int() with base 10: '4번째'
# 2번째 예외상황 : 인덱스에 없는 정수 입력한 경우 
    #  예외 종류: <class 'IndexError'>, 
    #  예외 객체: list index out of range


a = [1,2,3,4,5,6,7] # 인덱스범위 : 0~6

try:
  num = int(input("정수 입력>> "))
  print("{}번째 요소는 {}".format(num,a[num]))
except Exception as e:  # 예외 발생시키지 않고 e:예외 객체 
  print("예외 종류:", type(e))   
  print("예외 객체:", e)
```

```python
# 함수로 객체 만들기 
def create_player(name, p3,p2,rebound):
  return {"name" : name, 
        "point3":p3, 
        "point2": p2, 
        "rebound": rebound}
def score_sum(player):
   score_hap = player["point3"] + player["point2"]
   return score_hap 
   
players=[
      create_player("정대만",24,  7,  1),
      create_player("강백호", 0, 1,  10),
      create_player("서태웅", 12, 16,  5)]


for player in players: 
  print(player["name"],score_sum(player))
```
