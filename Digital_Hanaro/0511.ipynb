### 인덱스 & 인덱싱
1. 인덱스: 자료의 위치 (숫자:정수, 문자)

2. 인덱싱 : 인덱스를 이용하여 자료를 추출하는 것 
- print("hello")

```python
# 안녕하세요
# 1   2  3  4  5 => 원인데스 = 보통 숫자 세는 방식
# 0   1  2  3  4 => 제로인덱스 = 파이썬 숫자 세는 방식 
# -5 -4 -3 -2 -1
print("안녕하세요!"[0])
"qwert"[2]
```

### 슬라이싱
- 인덱스 범위 지정 
- [시작:끝]
- 끝번호를 포함하지 않음 


```python
print("s : ", s,"\n")

print("[0:4] =>", s[0:4])
print("[:4] =>", s[:4])
print("[:-3] =>", s[:-3])
print()
print("[3:] =>", s[3:])
print("[-4:] =>", s[-4:])
print()
print("[2:5] =>",s[2:5])
print("[-6:-2] =>",s[-6:-2])
```
