## 연결 vs 추가 

```python
# 연결 +
list1 = [1,2,3]
list2 = [4,5,6]

print(list1+list2)
```

```python
# 연결 *
print(list1*3)
```

```python
#리스트 길이 
len(list1)
```

```python
#추가 = 뒤에 추가 
list1.append(100)
print(list1)
```

```python
list1
```

```python
# insert(): 추가 중간에 추가 = 원하는 위치(인덱스 번호)
list1.insert(2,4) 
list1 # 파괴적 처리 
```

```python
#extend() : 추가 = + 연결과 같음 (원본의 변화도 같이 ) 리스트 전체 요소
list1.extend(list2)
print(list1)
print(list2)
```

## 리스트 요소 제거 
- del()
- pop()
- remove()

```python
# del
print(list1)
del list1[3]
print(list1)
```

```python
# pop 
print(list1)
list1.pop(3) # 3번째 인덱스 
print(list1)
```

```python
for i in character:
  print(i)
```

```python
character["school"]
```

```python
for i in character:
  if type(character[i]) == dict: # 키에 대한 value의 타입이 딕셔너리이면 
    for j in character[i]:
      print(character[i][j]) # 이중 딕셔너리의 value를 가지고 옴 
  elif type(character[i]) ==list: # 키에 대한 value의 타입이 리스트이면 
    for k in character[i]:# 리스트 안의 요소값들을 하나씩 가지고 옴 
      print(k)
  else: print(i)
```

```python
for i in character:  # 키를 가지고 옴 
  print(i)
```

```python
for i in character.values():
  print(i)
```

```python
for j, i in character.items():
  print(j,i)
```

## break / continue

### break

```python
i=0
while True:
  print(i)
  yn = input("종료를 원하면 Y OR y입력하세용~ >> ")
  if yn in ["y", "Y"]:
    print("조건 만족, 반복 종료")
    break
```

```python
i = 0
while True:    # 무한 루프
    print(i)
    i += 1          # i를 1씩 증가시킴
    if i == 10:    # i가 10일 때
        break       # 반복문을 끝냄. while의 제어흐름을 벗어남
```

### continue
- 현재 반복을 생략하고 다음 반복으로 

```python
list1 = [100,30,60,400,4,87,345]
for i in list1:
  if i<100:
    print(i)
  else: continue
```

```python
for i in range(20):       # 0부터 99까지 증가하면서 20번 반복
    if i % 2 == 0:         # i를 2로 나누었을 때 나머지가 0면 짝수
        continue           # 아래 코드를 실행하지 않고 건너뜀
    print(i)
```

#### 리스트 내포 

```python
list1 = []
list2 = []
for i in range(1,6):  
  list1.append(i)
  list2.append(i**2)
print(list1);print(list2)
```

```python
#list 내포로 재현 

list1 = [i for i in range(1,6)]
list2 = [i**2 for i in range(1,6)]

print(list1)
print(list2)
```

```python
result = [x * y for x in range(1, 5) for y in range(1, 5)] 
print(result)
```

```python
 # 리스트 내포 + 조건문 
 result1 = [x * y for x in range(1, 5) if x%2 == 0 for y in range(1, 5)] 
 print(result1)
```
