### 정수형(d)

```python
a = "{:d}".format(23) #정수
b = "{:5d}".format(23) # 자리수 & 남은자리를 빈칸으로 채움
c = "{:05d}".format(23) # 자리수 & 남은자리를으로 채움
d =  "{:05d}".format(-23) # 자리수 & 남은자리를으로 채움

print(a+"|"+b+"|"+c)
print("a:",a)
print("b:",b)
print("c:",c)
print("d:",d) # -도 차리를 차지 
```

```python
a1 = "{:+d}".format(23)  # 양수기호와 함께 출력
b1 = "{:-d}".format(23)  # d앞에 -기호는 의미 없음 
c1 = "{: d}".format(23)  # 양수기호 생략 출력
d1 = "{: d}".format(-23) # 음수, 무조건 -기호
e1 = "{:+d}".format(-23) # 음수, 무조건 -기호

print(a1 , b1,  c1, d1,e1)
```

```python
# a = "{:d}".format(1.23)
# ValueError: Unknown format code 'd' for object of type 'float'
```

```python
a = "{:+5d}".format(23)      #숫자 바로 앞에 기호 
b = "{:+5d}".format(-23)     
c = "{:=+5d}".format(23)     # 자리수 & 맨 앞에 기호 & 빈 자리를 빈칸으로 
d = "{:=+5d}".format(-23)    
e = "{:=+05d}".format(23)    # 자리수 & 맨 앞에 기호 & 빈 자리를 0으로 
f = "{:=+05d}".format(-23)  
g = "{:=05d}".format(23)     # 등호가 있지만 부호가 없는 경우: +는 생략 됨 
h = "{:=05d}".format(-23)    # 등호가 있지만 부호가 없는 경우: -는 생략 안 됨 

print("a:",a)
print("b:",b)
print("c:",c)
print("d:",d) # -도 차리를 차지 

print("e:",e,"f:",f,"g:",g,"h:",h)
```

```python
g = "{:=5d}".format(23)    # 
h = "{:=5d}".format(-23)  
print(g,h)
```

### 실수형 (f)

```python
a = "{:f}".format(23.456) #정수
b = "{:15f}".format(23.456) # 자리수 & 남은자리를 빈칸으로 채움
c = "{:+15f}".format(23.456) # 자리수 & 남은자리를으로 채움
c1 = "{:+15.5f}".format(23.456) # 자리수 & 남은자리를으로 채움
d =  "{:+015f}".format(-23.456) # 자리수 & 남은자리를으로 채움
d1 =  "{:+015.5f}".format(-23.456) # 자리수 & 남은자리를으로 채움


print(a+"|"+b+"|"+c)
print("a:",a)
print("b:",b)
print("c:",c)
print("c1:",c1)
print("d:",d) # -도 차리를 차지 
print("d1:",d1)
```

### ▶2-1) f-string과 왼쪽 정렬, 오른쪽 정렬, 가운데 정렬
- 아래 코드에서 보이듯이 중괄호 {}안에 있는 변수 뒤에 콜론(:)을 붙인 후 **왼쪽 정렬 (<), 오른쪽 정렬(>), 가운데 정렬(^)**의 옵션을 넣어줍니다. 그 후에 자릿수를 알려주는 숫자 를 넣어주면 정렬 옵션을 사용할 수 있다.

```python
# f-string 왼쪽 정렬
s1 = 'left'
result1 = f'|{s1:<10}|'
print(result1)
 
# f-string 가운데 정렬
s2 = 'mid'
result2 = f'|{s2:^10}|'
print(result2)
 
# f-string 오른쪽 정렬
s3 = 'right'
result3 = f'|{s3:>10}|'
print(result3)
```

### ▶2-2) f-string에서 중괄호 출력 방법
-  {{ }} 괄호를 연속해서 두개를 사용하면 중괄호 자체를 출력 할 수 있습니다.
그럼 이상황에서 f 스트링의 값과 중괄호 까지 같이 표현하려면 {{{, }}} 이렇게 중괄호를 세개 입력하면 num을 변수 값으로 인식할 수 있다.


```python
# f-string 중괄호 출력
num = 10
result = f'my age {{{num}}}, {{num}}'
print(result)
```

### ▶ 2-1 : 직접 대입, 변수로 대입, 이름으로 대입

```python
# 직접 대입하기
s1 = 'name : {0}'.format('BlockDMask')
print(s1)
 
# 변수로 대입 하기
age = 55
s2 = 'age : {0}'.format(age)
print(s2)
 
# 이름으로 대입하기
s3 = 'number : {num}, gender : {gen}'.format(num=1234, gen='남')
print(s3)
print(type(s3))
```

```python
s4 = 'name : {}, city : {}'.format('BlockDMask', 'seoul')
print(s4)
 
 
# 인덱스 순서가 바뀌면?
s5 = 'song1 : {1}, song2 : {0}'.format('nunu nana', 'ice cream')
print(s5)
 
 
# 인덱스를 중복해서 입력하면?
s6 = 'test1 : {0}, test2 : {1}, test3 : {0}'.format('인덱스0','인덱스1')
print(s6)
```
